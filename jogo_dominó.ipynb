```python
import random

def criar_pecas():
    pecas = []
    for i in range(7):
        for j in range(i, 7):
            pecas.append((i, j))
    return pecas

def embaralhar_pecas():
    pecas = criar_pecas()
    random.shuffle(pecas)
    return pecas

def distribuir_pecas(pecas):
    jogador1 = pecas[:7]
    jogador2 = pecas[7:14]
    pecas_mesa = pecas[14:]
    return jogador1, jogador2, pecas_mesa

def imprimir_mao(jogador):
    print("Suas peças:")
    for i, peca in enumerate(jogador):
        print(f"{i+1}: {peca}")

def verificar_peca(jogador, peca_escolhida):
    if peca_escolhida in jogador:
        return True
    return False

def verificar_jogada(mesa, peca_escolhida, lado):
    if lado == "comeco":
        #print(f"Debug: peca_escolhida[1]: {peca_escolhida[1]}")
        if peca_escolhida[1] == mesa[0][0] or peca_escolhida[0] == mesa[0][0]:
            return True
    elif lado == "final":
        #print(f"Debug: peca_escolhida[0]: {peca_escolhida[0]}")
        if peca_escolhida[1] == mesa[len(mesa)-1][1] or peca_escolhida[0] == mesa[len(mesa)-1][1]:
            return True
    return False

def jogar_peca(jogador, mesa, peca_escolhida, lado):
    print(f"debug: peca_escolhida: {peca_escolhida}")
    if lado == "comeco":
        if peca_escolhida[1] == mesa[0][0]:
            mesa.insert(0, peca_escolhida)
            #print(f" {peca_escolhida}inserido com sucesso")
            jogador.remove(peca_escolhida)
        else:
            jogador.remove(peca_escolhida)
            peca_escolhida = (peca_escolhida[1], peca_escolhida[0])
            mesa.insert(0, peca_escolhida)
            #print(f" {peca_escolhida}inserido com sucesso")

    elif lado == "final":
        if peca_escolhida[0] == mesa[len(mesa)-1][1]:
            mesa.append(peca_escolhida)
            #print(f" {peca_escolhida}inserido com sucesso")
            jogador.remove(peca_escolhida)
        else:
            jogador.remove(peca_escolhida)
            peca_escolhida = (peca_escolhida[1], peca_escolhida[0])
            mesa.append(peca_escolhida)
            #print(f" {peca_escolhida}inserido com sucesso")


def verificar_vitoria(jogador):
    if len(jogador) == 0:
        return True
    return False

def verificar_empate(jogadores, pecas_mesa):
    if len(jogadores[0]) == 0 and len(jogadores[1]) == 0 and len(pecas_mesa) == 0:
        return True
    return False

def decidir_lado_jogada(mesa, peca_escolhida):
    if peca_escolhida[0] == mesa[0][0] or peca_escolhida[1] == mesa[0][0]:
        return "comeco"
    elif peca_escolhida[0] == mesa[len(mesa)-1][1] or peca_escolhida[1] == mesa[len(mesa)-1][1]:
        return "final"
    return None

def iniciar_jogo():
    pecas = embaralhar_pecas()
    jogador1, jogador2, pecas_mesa = distribuir_pecas(pecas)
    mesa = []

    print("=== JOGO DE DOMINÓ ===\n")

    turno = random.randint(0, 1)
    jogadores = [jogador1, jogador2]
    pulou_vez = False

    mesa.append(pecas_mesa.pop())

    while True:
        jogador_atual = jogadores[turno]
        print("=== Jogador", turno + 1, "===\n")
        imprimir_mao(jogador_atual)
        print("Peças na mesa:", mesa)

        if pulou_vez:
            print("O jogador", turno + 1, "pulou a vez.\n")
            turno = (turno + 1) % 2
            pulou_vez = False
        else:
            opcao = input("Digite o número da peça que deseja jogar (ou '0' para pular): ")
            if opcao == '0':
                print("O jogador", turno + 1, "pulou a vez.\n")
                pulou_vez = True
            else:
                opcao = int(opcao)
                if opcao < 1 or opcao > len(jogador_atual):
                    print("Opção inválida.\n")
                else:
                    peca_escolhida = jogador_atual[opcao - 1]
                    if verificar_peca(jogador_atual, peca_escolhida):
                        lado = decidir_lado_jogada(mesa, peca_escolhida)
                        if lado is not None:
                            print(f"Debug: lado: {lado}")
                            if verificar_jogada(mesa, peca_escolhida, lado):
                                jogar_peca(jogador_atual, mesa, peca_escolhida, lado)
                                print(f"Debug: jogador: {jogador_atual} \n")
                                print(f"Debug: mesa: {mesa} \n")
                                print(f"Debug: escolha: {peca_escolhida}")
                                print(f"Debug: lado: {lado} \n")
                                if verificar_vitoria(jogador_atual):
                                    print("Jogador", turno + 1, "venceu!")
                                    break
                                elif verificar_empate(jogadores, pecas_mesa):
                                    print("Empate!")
                                    break
                                turno = (turno + 1) % 2
                            else:
                                print("Jogada inválida.\n")
                        else:
                            print("Não é possível jogar essa peça.\n")
                    else:
                        print("Peça inválida.\n")

iniciar_jogo()
```
